```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from ipywidgets import widgets
from IPython.display import display

# Load the dataset
dataset = pd.read_csv('Crop and fertilizer dataset.csv')

# Create widgets
initial_district = dataset['District_Name'].unique()[0]
initial_soil_colors = dataset[dataset['District_Name'] == initial_district]['Soil_color'].unique()

district_widget = widgets.Dropdown(options=dataset['District_Name'].unique(), description='District', value=initial_district)
soil_color_widget = widgets.Dropdown(options=[(color, color) for color in initial_soil_colors], description='Soil Color')
nitrogen_widget = widgets.FloatText(description='Nitrogen')
phosphorus_widget = widgets.FloatText(description='Phosphorus')
potassium_widget = widgets.FloatText(description='Potassium')
ph_widget = widgets.FloatText(description='pH')
rainfall_widget = widgets.FloatText(description='Rainfall')
temperature_widget = widgets.FloatText(description='Temperature')
recommend_widget = widgets.Output()

# Function to update soil color options when district changes
def update_soil_color_options(change):
    district = district_widget.value
    if district:
        soil_colors = dataset[dataset['District_Name'] == district]['Soil_color'].unique()
        soil_color_widget.options = [(color, color) for color in soil_colors]
    else:
        soil_color_widget.options = []

# Define the observer function to update nutrient and environmental factors
def update_values(change):
    district = district_widget.value
    soil_color = soil_color_widget.value
    if district and soil_color:
        nitrogen_values = dataset[(dataset['District_Name'] == district) & (dataset['Soil_color'] == soil_color)]['Nitrogen'].unique()
        phosphorus_values = dataset[(dataset['District_Name'] == district) & (dataset['Soil_color'] == soil_color)]['Phosphorus'].unique()
        potassium_values = dataset[(dataset['District_Name'] == district) & (dataset['Soil_color'] == soil_color)]['Potassium'].unique()
        ph_values = dataset[(dataset['District_Name'] == district) & (dataset['Soil_color'] == soil_color)]['pH'].unique()
        rainfall_values = dataset[(dataset['District_Name'] == district) & (dataset['Soil_color'] == soil_color)]['Rainfall'].unique()
        temperature_values = dataset[(dataset['District_Name'] == district) & (dataset['Soil_color'] == soil_color)]['Temperature'].unique()
        
        nitrogen_widget.value = nitrogen_values[0] if len(nitrogen_values) > 0 else 0
        phosphorus_widget.value = phosphorus_values[0] if len(phosphorus_values) > 0 else 0
        potassium_widget.value = potassium_values[0] if len(potassium_values) > 0 else 0
        ph_widget.value = ph_values[0] if len(ph_values) > 0 else 0
        rainfall_widget.value = rainfall_values[0] if len(rainfall_values) > 0 else 0
        temperature_widget.value = temperature_values[0] if len(temperature_values) > 0 else 0

# Function to train model and display predictions
def train_model(b):
    # Get the selected values from the widgets
    district = district_widget.value
    soil_color = soil_color_widget.value
    nitrogen = nitrogen_widget.value
    phosphorus = phosphorus_widget.value
    potassium = potassium_widget.value
    pH = ph_widget.value
    rainfall = rainfall_widget.value
    temperature = temperature_widget.value

    if not district or not soil_color:
        with recommend_widget:
            recommend_widget.clear_output()
            print("Please select a district and soil color.")
            return

    # Create input data
    input_data = pd.DataFrame([[nitrogen, phosphorus, potassium, pH, rainfall, temperature, district, soil_color]],
                              columns=['Nitrogen', 'Phosphorus', 'Potassium', 'pH', 'Rainfall', 'Temperature', 'District_Name', 'Soil_color'])

    # Perform One-Hot Encoding for District_Name and Soil_color columns
    encoder = OneHotEncoder(handle_unknown='ignore')
    X_encoded = encoder.fit_transform(dataset[['District_Name', 'Soil_color']])
    input_data_encoded = encoder.transform(input_data[['District_Name', 'Soil_color']])

    # Train a simple model (RandomForestClassifier) on the dataset
    model_crop = RandomForestClassifier()
    model_crop.fit(X_encoded, dataset['Crop'])

    # Make predictions using the trained model
    predicted_crop = model_crop.predict(input_data_encoded)

    # Find the fertilizer associated with the recommended crop
    recommended_fertilizer = dataset[dataset['Crop'] == predicted_crop[0]]['Fertilizer'].values[0]

    with recommend_widget:
        recommend_widget.clear_output()
        print("Recommended Crop: ", predicted_crop[0])
        print("Recommended Fertilizer: ", recommended_fertilizer)

# Set up observers
district_widget.observe(update_soil_color_options, names='value')
soil_color_widget.observe(update_values, names='value')

# Create the button widget
button = widgets.Button(description='Recommend Crop')
button.on_click(train_model)

# Display the widgets
display(widgets.VBox([
    district_widget, soil_color_widget, nitrogen_widget, phosphorus_widget,
    potassium_widget, ph_widget, rainfall_widget, temperature_widget,
    button, recommend_widget
]))
```
